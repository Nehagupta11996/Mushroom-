```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaledX = scaler.fit_transform(df)
```

```python
scaledX
```

standard scaler is done to make mean as 0 and standard deviation as 1
this is the prerequisite of PCA the data should be scalled

```python
from sklearn.decomposition import PCA

testpca=PCA()
Y=testpca.fit(scaledX)
```

```python
Y.explained_variance_ratio_
```

```python
var_cum = np.cumsum(Y.explained_variance_ratio_)*100
```

```python
var_cum
```

This gives the information of the data that by taking how many columns and how many rows we can get maximum information.


```python
k=np.argmax(var_cum>95)
print("no of components explaining 94% variance:",k)
```

```python
FinalPCA=PCA(n_components=15)
FinalData = FinalPCA.fit_transform(scaledX)
```

```python
for i in range(200):
    x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.33, random_state=i)
    from sklearn.tree import DecisionTreeClassifier
    lm=DecisionTreeClassifier()
    lm.fit(x_train,y_train)
    pred=lm.predict(x_test)
    accu=accuracy_score(y_test,pred)
    print(accu*100, "at random_state",i)
```
